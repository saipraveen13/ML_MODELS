```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df =  pd.read_csv("/content/CarPrice_Assignment.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Age'] = df['car_ID'].max() - df['car_ID']
```

```python
X = df[['Age']]
y = df['price']
```

```python
plt.scatter(X, y, color='blue')
plt.xlabel("Car Age (simulated)")
plt.ylabel("Car Price")
plt.title("Car Price vs Age")
plt.show()
```

```python
plt.boxplot(df['Age'])
plt.show()
```

```python
# Generate polynomial features
poly_features = PolynomialFeatures(degree=2) # You can change the degree as needed
X_poly = poly_features.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(X_poly, y)
```

```python
y_pred = model.predict(X_poly)
```

```python
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel("Car Age (simulated)")
plt.ylabel("Car Price")
plt.title("Polynomial Regression Fit (Car Price vs Age)")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", np.sqrt(mse))
print("RÂ² Score:", r2)
```
