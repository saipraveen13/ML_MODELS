```python
df['Pass_Fail'] = df['GRADE'].apply(lambda x: 0 if x < 2 else 1)
```

```python
features = df.drop(columns=['STUDENT ID', 'COURSE ID', 'GRADE', 'Pass_Fail'])
target = df['Pass_Fail']
```

```python
features = pd.get_dummies(features, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, stratify=target)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python

features = df.drop(columns=['STUDENT ID', 'COURSE ID', 'GRADE', 'Pass_Fail'])
target = df['Pass_Fail']

pass_indices = target[target == 1].index
fail_indices = target[target == 0].index

num_fail_train = max(1, int(len(fail_indices) * 0.7))
num_pass_train = int(len(pass_indices) * 0.7)

import numpy as np
np.random.seed(42)
train_fail_indices = np.random.choice(fail_indices, num_fail_train, replace=False)
train_pass_indices = np.random.choice(pass_indices, num_pass_train, replace=False)

train_indices = np.concatenate([train_fail_indices, train_pass_indices])
test_indices = df.index.difference(train_indices)

X_train = features.loc[train_indices]
y_train = target.loc[train_indices]
X_test = features.loc[test_indices]
y_test = target.loc[test_indices]

print("Value counts in y_train after manual split:")
print(y_train.value_counts())
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
